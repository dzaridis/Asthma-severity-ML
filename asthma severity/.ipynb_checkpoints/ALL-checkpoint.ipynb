```python
def curves(y_test,y_score_DT,y_pred_DT):
    sens_DT=np.sum((y_score_DT==1) & (y_test==1))/ np.sum(y_test)
    [TN, FP, FN ,TP]=confusion_matrix(y_test,y_score_DT).ravel()
    sens_DT=TP/(TP+FN)
    spec_DT=TN/(TN+FP)
    fpr_DT, tpr_DT, thresholds = roc_curve(y_test,y_pred_DT)
    acc_DT=accuracy_score(y_test,y_pred_DT)
    recall_DT=recall_score(y_test,y_pred_DT,average='macro')
    prec_DT=precision_score(y_test,y_pred_DT)
    roc_auc=auc(fpr_DT,tpr_DT) 
    return [fpr_DT,tpr_DT,sens_DT,spec_DT,acc_DT,recall_DT,prec_DT,roc_auc,TN,FP,FN,TN];
```
